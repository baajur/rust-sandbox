```python
%matplotlib inline
import csv
i = 0
degrees = []
with open("log.tsv", newline="") as f:
    reader = csv.reader(f, delimiter = '\t')
    for row in reader:
        if i % 10 == 0:
            degrees.append(float(row[3])/180)
        i += 1
len(degrees)
```

```python
import matplotlib.pyplot as plt
plt.plot(degrees)
plt.ylabel('some numbers')
plt.show()
```
