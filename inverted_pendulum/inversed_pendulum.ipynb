```python
%matplotlib inline
import csv
i = 0
x_lst = []
v_lst = []
theta_lst = []
omega_lst = []
with open("inverted_pendulum.tsv", newline="") as f:
    reader = csv.reader(f, delimiter = '\t')
    for row in list(reader)[0:100000]:
        if i % 1 == 0:
            x_lst.append(float(row[1]))
            v_lst.append(float(row[2]))
            theta_lst.append(float(row[4]))
            omega_lst.append(float(row[5]))
        i += 1
len(x_lst)
```

```python
import matplotlib.pyplot as plt
plt.plot(x_lst)
plt.ylabel('some numbers')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(v_lst)
plt.ylabel('some numbers')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(theta_lst)
plt.ylabel('some numbers')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.plot(omega_lst)
plt.ylabel('some numbers')
plt.show()
```
